# 1. 주요 활성화 함수(Activation Function)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-10.0,10.0,0.1)
```

<br>

## 1.1 계단 함수 (Step Function)

$$
y = 
\begin{cases}
    1, & \text{if  } W X + B > 0 \\
    0, & \text{if  } W X + B \leq 0
\end{cases}
$$

```python
def stepFunction(x):
    return np.array(x > 0, dtype=np.int)
```

```python
y = stepFunction(x)

plt.plot(x, y)
plt.title("Step Function")
plt.show()
```

<br>

## 1.2 시그모이드 함수 (Sigmoid Function)

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

```python
def sigmoidFunction(x):
    return 1 / (1 + np.exp(-x))
```

```python
y = sigmoidFunction(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.show()
```

<br>

## 1.3 tanh 함수 (Hyperbolic tangent function, 하이퍼볼릭 탄젠트 함수)

$$
tanh(x) = 2 \sigma(2x) - 1
$$

$$
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

```python
def tanhFunction(x):
    return np.tanh(x)
```

```python
y = tanhFunction(x)

plt.plot(x, y)
plt.title("tanh Function")
plt.show()
```

<br>

## 1.4 렐루 함수 (ReLU)

$$
f(x) = max(0, x)
$$

```python
def ReLUFunction(x):
    return np.maximum(0, x)
```

```python
y = ReLUFunction(x)

plt.plot(x, y)
plt.title("ReLU Function")
plt.show()
```

<br>

## 1.5 리키 렐루 함수 (Leaky ReLU Function)

$$
f(x) = max(a x, x) \quad \text{a : hyper parameter}
$$

```python
def leakyReLUFunction(x, a=0.01):
    return np.maximum(a*x, x)
```

```python
y = leakyReLUFunction(x, a=0.1)

plt.plot(x, y)
plt.title("Leaky ReLU Function")
plt.show()
```

<br>

## 1.6 소프트맥스 함수 (Softmax Function)

$$
f(x) = \frac{e^{x_k}}{\sum_{i=1}^n e^{x_i}}
$$

```python
def softmaxFunction(x):
    return np.exp(x)/np.sum(np.exp(x))
```

```python
y = softmaxFunction(x)

plt.plot(x, y)
plt.title("Softmax Function")
plt.show()
```
